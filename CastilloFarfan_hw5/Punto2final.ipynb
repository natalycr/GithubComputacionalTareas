#### Se obtubo de  dos formas: la primera  con fft y la  segunda  con rfft 

# 1) Usando fft

```python
figsize(15,7)
#  transformadas de la senal original 

fft_original=np.fft.fft(input_signal)
w=np.fft.fftfreq(len(fft_original),1)

print "fft_original=", len(fft_original), 'w=', len(w)

plt.subplot(121),plot(np.abs(fft_original)), plt.title('Transformada de la senal Original')
plt.subplot(122),plot(log10(np.abs(fft_original)**2)), plt.title('Espectro de frecuencias senal Original')
```

```python
scatter(w,fft_original)
xlabel("Frecuencias", fontsize=20)
ylabel("Transformada", fontsize=20)
```

```python
ruidoBlanco_fft = fft_original/np.abs(fft_original)
print "tamano ruiso norm", len(ruidoBlanco_fft) #, ruidoBlanco_fft

#plt.subplot(121), plot(np.abs(ruidoBlanco_fft))
plt.subplot(121), plot(w, np.abs(ruidoBlanco_fft)), plt.title('Ruido en el espacio de fourier')
xlabel("Frecuencias", fontsize=20)
ylabel("Transformada", fontsize=20)

elabs=(ruidoBlanco_fft)**2 

plt.subplot(122),plot(w, log10(elabs)), plt.title('Espectro de frecuencias')
xlabel("Frecuencias", fontsize=20)
ylabel("|1/Transformada|^2", fontsize=20)
```

##### Nota: Estan  dando al reves. 

```python
ifft_RuidoBlanco= np.fft.ifft(ruidoBlanco_fft)

print 'tiempo',len(time_array),'w', len(w), 'senal', len(ifft_RuidoBlanco), 'normaliz', len(ruidoBlanco_fft)

plt.plot(time_array, ifft_RuidoBlanco), plt.title('Ruido Blanco')

# cre el  archivo de audio
a=real(ifft_RuidoBlanco)
output_file = audio.Sndfile("ruidoblanco.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(a)
output_file.close()
```

```python
#  Ruido Rosa 
#=decaimiento 1/f de  forma  exponencial 
funex=np.exp(w)
print funex
rosa1_fft=ruidoBlanco_fft/(funex)
#print 'rosa1', len(rosa1_fft), rosa1_fft
#plt.subplot(121),plot(w, rosa1_fft) # Daveria ser exponencial decreciente 
es=(rosa1_fft)**2
plot(w, log10(es)), plt.title('Espectro de frecuencias') # deberia ser exponencial 
```

```python
ifft_rosa= np.fft.ifft(rosa1_fft)
plt.plot(ifft_rosa), plt.title('Ruido Rosa')
print ifft_rosa

# Archivo  de Ruido Rosa 
RealRosa=real(ifft_rosa)
output_file = audio.Sndfile("ruidoRosa.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(RealRosa)
output_file.close()
```

## 2) Usando ifft

```python
fft_original=np.fft.ifft(input_signal)
w=np.fft.fftfreq(len(fft_original),1)

#print "fft_original=", len(fft_original), 'w=', len(w)

plt.subplot(121),plot(np.abs(fft_original)), plt.title('Transformada de la senal Original')
plt.subplot(122),plot(log10(np.abs(fft_original)**2)), plt.title('Espectro de frecuencias senal Original')
# se ven alreves 
```

```python
ruidoBlanco_fft = fft_original/np.abs(fft_original)
print "tamano ruiso norm", len(ruidoBlanco_fft) #, ruidoBlanco_fft

#plt.subplot(121), plot(np.abs(ruidoBlanco_fft))
plt.subplot(121), plot(w, np.abs(ruidoBlanco_fft)), plt.title('Ruido en el espacio de fourier')
xlabel("Frecuencias", fontsize=20)
ylabel("Transformada", fontsize=20)

elabs=(ruidoBlanco_fft)**2 

plt.subplot(122),plot(w, log10(elabs)), plt.title('Espectro de frecuencias')
xlabel("Frecuencias", fontsize=20)
ylabel("|1/Transformada|^2", fontsize=20)
```

##### Nota: Tambien  se  ve al reves 

```python
ifft_RuidoBlanco= np.fft.irfft(ruidoBlanco_fft)

print 'tiempo',len(time_array),'w', len(w), 'senal', len(ifft_RuidoBlanco), 'normaliz', len(ruidoBlanco_fft)

plt.plot(time_array, ifft_RuidoBlanco), plt.title('Ruido Blanco')

# cre el  archivo de audio
output_file = audio.Sndfile("ruidoblanco2.wav", 'w', audio.Format('wav'), 1, sampling_rate)
output_file.write_frames(ifft_RuidoBlanco)
output_file.close()
```
